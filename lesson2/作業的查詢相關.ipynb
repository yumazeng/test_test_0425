請用python寫一段程式
1.請讀取names.txt
2.將文字儲存於list內
3.每次執行可以亂數取出3個名字

當然可以！以下是一段用 Python 實作的程式碼，會讀取 `names.txt` 檔案，將每一行的名字存入一個 list，並且每次執行時隨機抽出 3 個名字：

```python
import random

# 讀取 names.txt，將每行名字存入 list
with open('names.txt', 'r', encoding='utf-8') as file:
    names = [line.strip() for line in file if line.strip()]  # 去除空行與換行符號

# 隨機抽取 3 個不重複的名字
random_names = random.sample(names, 3)

print("隨機抽出的三個名字：")
for name in random_names:
    print(name)
```

## 說明
- `with open(...)` 以安全方式開啟檔案，並用 `utf-8` 編碼讀取。
- 利用 list comprehension 去除空白行。
- `random.sample()` 用來從 list 中抽取指定數量的不重複元素。
- 如果 `names.txt` 中名字少於 3 個，`random.sample()` 會丟出錯誤，可以再加判斷避免。

請加上錯誤處理

當然可以！以下是加上錯誤處理的完整範例，包含：

- 檔案不存在時的錯誤提示
- 檔案中名字不足 3 個時的提示
- 其他可能的例外捕捉

```python
import random

try:
    with open('names.txt', 'r', encoding='utf-8') as file:
        names = [line.strip() for line in file if line.strip()]

    if len(names) < 3:
        print(f"錯誤：檔案中只有 {len(names)} 個名字，無法抽出 3 個。")
    else:
        random_names = random.sample(names, 3)
        print("隨機抽出的三個名字：")
        for name in random_names:
            print(name)

except FileNotFoundError:
    print("錯誤：找不到檔案 'names.txt'，請確認檔案是否存在於程式所在目錄。")

except Exception as e:
    print(f"發生未知錯誤：{e}")
```

---

## 說明

- `try` 區塊包裹主要程式碼，讓我們可以捕捉並處理執行過程中可能發生的錯誤。
- `FileNotFoundError`：當 `names.txt` 不存在時觸發，會給出友善的錯誤訊息。
- `if len(names) < 3`：檢查檔案中名字數量是否足夠，避免 `random.sample()` 拋出錯誤。
- `except Exception as e`：捕捉其他所有未知錯誤，並印出錯誤訊息，方便除錯。
