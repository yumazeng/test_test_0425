## Q:python list 的實體方法 insert

## A:
在 Python 中，`list.insert()` 是一個**實體方法**（也稱為 instance method），用來在指定位置**插入元素**到清單中，而不會覆蓋原有的元素。

### 語法：
```python
list.insert(index, element)
```

### 參數說明：
- `index`：要插入元素的位置索引。如果 `index` 大於列表長度，則會將元素插入到列表末尾；如果是負數，則從列表尾部反向計算。
- `element`：要插入的元素。

### 回傳值：
- 此方法 **不會回傳任何值**，也就是說它會 **直接修改原始的 list**（in-place 操作），並回傳 `None`。

---

### 範例：
```python
fruits = ['apple', 'banana', 'cherry']
fruits.insert(1, 'orange')
print(fruits)
```

#### 輸出：
```
['apple', 'orange', 'banana', 'cherry']
```

---

### 特殊情況：
```python
nums = [1, 2, 3]
nums.insert(100, 99)  # 索引超出範圍，會加到最後
print(nums)  # [1, 2, 3, 99]

nums.insert(-1, 0)  # 負索引，插入倒數第2個位置
print(nums)  # [1, 2, 0, 3, 99]
```


Q:python list 的實體方法 insert()如何使用

## A:
`list.insert()` 是 Python 中用來「在特定位置插入元素」的方法，它屬於 list 的實體方法，也就是只能透過某個 list 物件來呼叫。

---

### ✅ 使用方式（語法）：

```python
list_name.insert(index, element)
```

- `index`：你要插入元素的位置（整數）。
- `element`：你要插入的值（可以是任何型別）。

---

### 🎯 實例說明：

```python
numbers = [10, 20, 30]

# 在索引 1 的位置插入 15
numbers.insert(1, 15)

print(numbers)
```

#### 🔽 輸出：
```
[10, 15, 20, 30]
```

---

### 🧠 注意事項：

1. **不會取代原有元素**，而是把元素往右推。
2. **超出長度的 index**：會直接加到 list 的最後。
3. **負數 index**：代表從右邊開始算。

---

### 🧪 更多範例：

```python
letters = ['a', 'b', 'c']

letters.insert(0, 'z')     # 插入到最前面
print(letters)             # ['z', 'a', 'b', 'c']

letters.insert(10, 'x')    # 插入到最後（因為 index 太大）
print(letters)             # ['z', 'a', 'b', 'c', 'x']

letters.insert(-1, 'y')    # 插入到倒數第二的位置
print(letters)             # ['z', 'a', 'b', 'y', 'c', 'x']
```


---

### 🔸 範例 1：基本用法

```python
fruits = ['apple', 'banana', 'cherry']
fruits.insert(1, 'orange')
print(fruits)
```

📌 結果：
```
['apple', 'orange', 'banana', 'cherry']
```
> `orange` 被插入到索引 1 的位置，其他元素自動向後移動。

---

### 🔸 範例 2：插入到開頭或結尾

```python
numbers = [2, 3, 4]

numbers.insert(0, 1)    # 插入到最前面
numbers.insert(100, 5)  # 插入到最後，因為 index 超過長度

print(numbers)
```

📌 結果：
```
[1, 2, 3, 4, 5]
```

---

### 🔸 範例 3：負數索引

```python
letters = ['a', 'c', 'd']
letters.insert(-1, 'b')  # 插入到倒數第2個位置

print(letters)
```

📌 結果：
```
['a', 'c', 'b', 'd']
```

## 🧩 `insert()` 是什麼？

`insert()` 是用來**在清單（list）中插入一個新元素**的方法。  
它可以讓你指定要把元素放在哪個位置。

---

## 🛠️ 語法是這樣：

```python
清單名稱.insert(位置索引, 要插入的東西)
```

- `位置索引`：你想要插入的位置（從 0 開始算）。
- `要插入的東西`：任何你想加進去的資料，例如數字、文字、甚至另一個 list。

---

## ✅ 範例 1：基本用法

```python
水果 = ['蘋果', '香蕉', '芒果']

# 把「葡萄」插入到第 1 個位置（也就是香蕉前面）
水果.insert(1, '葡萄')

print(水果)
```

🔸 輸出結果：
```
['蘋果', '葡萄', '香蕉', '芒果']
```

---

## ✅ 範例 2：插入到開頭或結尾

```python
數字 = [2, 3, 4]

數字.insert(0, 1)     # 插入到最前面
數字.insert(100, 5)   # 插入到最後（因為位置超過 list 長度）

print(數字)
```

🔸 結果：
```
[1, 2, 3, 4, 5]
```

---

## ✅ 範例 3：使用負數索引

```python
字母 = ['A', 'C', 'D']

字母.insert(-1, 'B')  # 插入到倒數第 2 個位置

print(字母)
```

🔸 結果：
```
['A', 'C', 'B', 'D']
```

---

## 🎯 小提醒：

- `insert()` 會**修改原本的清單**。
- 插入後，原本位置的元素會自動往後移。
- 不會回傳新清單，只會直接改變原來的 list。



Q: Python list 的實體方法 pop()



## 🧩 `pop()` 是什麼？

`pop()` 是用來**從清單中移除一個元素，並把它取出來**的方法。

---

## 🛠️ 使用語法：

```python
清單名稱.pop()
```

或

```python
清單名稱.pop(索引)
```

- **不加索引**：會移除清單中**最後一個**元素。
- **加索引**：會移除**指定位置的**元素。

🔸 而且 `pop()` **會回傳被刪除的值**，你可以把它存下來。

---

## ✅ 範例 1：移除最後一個元素

```python
水果 = ['蘋果', '香蕉', '芒果']
拿掉的 = 水果.pop()

print(拿掉的)   # 芒果
print(水果)     # ['蘋果', '香蕉']
```

---

## ✅ 範例 2：移除指定位置的元素

```python
水果 = ['蘋果', '香蕉', '芒果']
拿掉的 = 水果.pop(0)  # 移除第 0 個，也就是「蘋果」

print(拿掉的)   # 蘋果
print(水果)     # ['香蕉', '芒果']
```

---

## 🚨 注意：

- 如果你 `pop()` 時，清單是空的，會出現錯誤（`IndexError`）。
  
```python
空清單 = []
空清單.pop()  # 會報錯：IndexError: pop from empty list
```

---

## 📌 總結：

| 方法 | 作用 | 備註 |
|------|------|------|
| `pop()` | 移除並回傳最後一個元素 | 不加索引時用 |
| `pop(index)` | 移除並回傳指定位置的元素 | 索引從 0 開始 |
